```python
import pandas as pd
import matplotlib.pyplot as plt

bloques = pd.read_csv("db/bloques_mina.csv")
plan = pd.read_csv("db/planificacion.csv")
planta = pd.read_csv("db/planta_lixiviacion.csv")
```

```python
plt.hist(bloques["ley_au_estimado"], bins=20)
plt.title("Distribución ley de oro")
plt.xlabel("g/t")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.hist(plan["margen"], bins=30)
plt.title("Distribución de margen")
plt.xlabel("Margen (USD/t)")
plt.ylabel("Número de bloques")
plt.show()
```

```python
plt.plot(planta["periodo"], planta["ley_cabeza_au"])
plt.title("Ley de cabeza por periodo")
plt.xlabel("Periodo")
plt.ylabel("g/t")
plt.show()
```
